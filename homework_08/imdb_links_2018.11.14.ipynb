```python
import pandas as pd
import imdb
from bs4 import BeautifulSoup
import requests
import re
import json

my_movies = ["The Big Lebowski","Fargo"]
```

```python
tbl_imdb  = pd.DataFrame(columns = [
    'Movie_Name',
    'Review_Title',
    'Review_Desc',
    'Review_url'])

i_index=[]
i_index = 0
```

```python
for movie_name in my_movies:
    
    print(movie_name)
    
    ia = imdb.IMDb()
    results = ia.search_movie(movie_name)
    mv = results[0] #First result
    url = ia.get_imdbURL(mv) #URL for first result
    user_review_url = url+'reviews/'
    response = requests.get(user_review_url)
    
    soup = BeautifulSoup(response.text, 'lxml')
    
    try:
#        print("+++================== Begin Soup =================================+++\n")
#        print(soup)
#        print("+++================== End Soup =================================+++\n")

        movies = soup.select('a.title')
 #       print("+++================== Begin Movies =================================+++\n")
 #       print("movies", movies)
 #       print("+++================== End Movies =================================+++\n")

        links = [a.attrs.get('href') for a in soup.select('a.title')]
        
 #       print(links)
        # Store each item into dictionary (data), then put those into a list (imdb)
        
        for index in range(0, len(movies)):
            
            # Separate movie into: 'place', 'title', 'year'
            
            review_title = movies[index].get_text()
            
            review_desc = ""
            
#            print(index, " ", review_title)
            
            raw_data = {'Movie_Name' : movie_name,
                    'Review_Title' : review_title,
                    'Review_Desc' : review_desc,
                    'Review_url': links[index] }
            
            df_tbl = pd.DataFrame(raw_data,
                        columns = [
                        'Movie_Name',
                        'Review_Title',
                        'Review_Desc',
                        'Review_url'], index = [i_index + 1])
            
            tbl_imdb = tbl_imdb.append(df_tbl)
            
    except Exception as e:
        print('Error : ' + str(e))
        
```
