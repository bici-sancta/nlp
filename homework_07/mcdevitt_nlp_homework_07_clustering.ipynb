### Data preparation, feature construction     

- 100 movie reviews were downloaded from the IMDB web-site  
- the reviews (approx 25 reviews each) are associated to four different films :  
    * Stranger Than Paradise, Fargo, The Big Lebowski, and Small Apartments  
        - these four films were chosen from the base category drama  
        - each of these films has a certain cult following for its unique quirkiness  
        - two of these films (Lebowski and Fargo) are Coen Brothers films  
- the text from each review was normalized in the following method :   
    * contractions expanded  
    * part of speech tagged  
    * lemmatized, using the part of speech tags in WordNet  
    * special characters removed  
    * stop words removed, using NLTK english stop words  
        - a small list of additional stop words were added to base list, to remove terms that are common in film reviews, e.g., film, movie, cinema, scene, and similar common terms  
    * for one sequence of clusters developed here, an additional part of speech tagging was completed, and only Noun and Verb forms of tokens were retained - this was done to evaluate if using limited parts of speech in the clustered corpus produces more distinct cluster key features  
    * features were developed using TF-IDF feature matrix  
    * the feature matrix was used for non-supervised learning using k-means clustering method

```python
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... movie list to extract reviews
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

movie_list = ["Stranger Than Paradise", "The Big Lebowski", "Fargo", "Small Apartments"]

# ... abbreviations for each film -- use in summary data frame

movie_list_abbr = ['STP', 'TBL', 'FAR', 'SAP']

BASE_URL = "https://www.imdb.com"
```

```python
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... create table for collecting results
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

df_imdb  = pd.DataFrame(columns = [
    'movie_name',
    'review_title',
    'review_desc',
    'review_rating',
    'review_url'])

i_index=[]
i_index = 0
```

```python
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... query IMDB for list of reviews
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

for movie_name in movie_list:
    
    print(movie_name)
    
    ia = imdb.IMDb()
    
    results = ia.search_movie(movie_name)
    mv = results[0]
    
    url = ia.get_imdbURL(mv)
    
    user_review_url = url + 'reviews/'
    
    print(user_review_url)
    
    response = requests.get(user_review_url)
#    print(response.text)
    
    soup = BeautifulSoup(response.text, 'lxml')
    
    try:
        reviews = soup.select('a.title')
        links = [a.attrs.get('href') for a in soup.select('a.title')]
        
#        print("reviews list length = ", len(reviews))

        for index in range(0, len(reviews)):
            
            review_title = reviews[index].get_text()
            
            review_desc = "none"
            review_rating = "none"
            review_url = BASE_URL + links[index]
            r_response = requests.get(review_url)
            r_soup = BeautifulSoup(r_response.text, 'lxml')
            
            try:

                d1 = r_soup.find('script', type = 'application/ld+json').text
                data = json.loads(d1)
                                
                try:
                    review_desc = data['reviewBody']
                    
                except Exception as e :
 #                   print("Error review extract : " + str(e))
                    review_desc = "NA"
                    
                try:
                    review_rating = float(data['reviewRating']['ratingValue'])
                    
                except Exception as e :
 #                   print("Error review rating : " + str(e))
                    review_rating = float('nan')
                            
            except Exception as e :
                continue
 #               print("Error text extraction : " + str(e))
                
            raw_data = {'movie_name' : movie_name,
                    'review_title' : review_title,
                    'review_desc' : review_desc,
                    'review_rating' : review_rating,
                    'review_url' : links[index]}

            df_tbl = pd.DataFrame(raw_data,
                    columns = [
                        'movie_name',
                        'review_title',
                        'review_desc',
                        'review_rating',
                        'review_url'], index = [i_index + 1])

            df_imdb = df_imdb.append(df_tbl)
            
    except Exception as e:
        continue
#        print("Error acquire links : " + str(e))

# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... reset index of data frame 
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

df_imdb.reset_index(drop = True, inplace = True)
```

```python
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... Clustering
# ...
# ... methods and code below are based on examples as shown in
# ... Text Analytics with Python, Sarkar, pp 300 - 310
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... drop any rows where there is no text in review
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

df_imdb_subset = df_imdb[df_imdb.review_desc != 'NA']
corpus = df_imdb_subset['review_desc']
```

```python
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-
# ... cluster evaluation
# ... - 2 loops
# ...   - normalization method (retain all POS or retain only Noun and Verb forms)
# ...   - number of clusters from 2 to 6
# ... -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

norm = ['Retain all parts of speech', 'Retain only Noun & Verb forms']

for normalization in range(2) :
    
# ...  choose normalization method

    if(normalization == 0) :
        norm_reviews = normalize_corpus(corpus)   # all POS
    else :
        norm_reviews = normalize_corpus_NV(corpus)  # only Noun & Verb tokens retained
        
    tfidf_vectorizer, tfidf_feature_matrix = build_feature_matrix(norm_reviews,
                                                            feature_type = 'tfidf',
                                                            ngram_range = (1, 2), 
                                                            min_df = 0.1,
                                                            max_df = 0.9)

    feature_names = tfidf_vectorizer.get_feature_names()

# ... vary cluster size from 2 to 6

    for num_clusters in range(2,7) :

        df_cluster = df_imdb_subset

        print '-='*40
        print 'Normalization Method = %s' % (norm[normalization])
        print 'Number of Clusers = %d\n' % (num_clusters)
        print '-='*40

        km_obj, clusters = k_means(feature_matrix = tfidf_feature_matrix, num_clusters = num_clusters)

        df_cluster['cluster'] = clusters

        cluster_data =  get_cluster_data(clustering_obj = km_obj,
                                         df_to_cluster = df_cluster,
                                         feature_names = feature_names,
                                         num_clusters = num_clusters,
                                         topn_features = 20)
        
#        print_cluster_data(cluster_data)

        df_clstr =  make_cluster_agg(clustering_obj = km_obj,
                                            df_to_cluster = df_cluster,
                                            feature_names = feature_names,
                                            num_clusters = num_clusters,
                                            topn_features = 10,
                                            movie_list = movie_list,
                                            movie_list_abbr = movie_list_abbr)
        df_clstr
        print '-='*40
        print '\n'

# ... save some subsets for results summary
        
        if (num_clusters == 4) :
            if(normalization == 0):
                df_clstr_sub1 = df_clstr.copy()
                del df_clstr_sub1['key_features']
#                print tabulate(df_clstr_sub1, headers='keys', tablefmt='psql')
                df_clstr_sub2 = df_clstr['key_features'].copy()
            if(normalization == 1):
                df_clstr_sub3 = df_clstr.copy()
                del df_clstr_sub3['key_features']
                df_clstr_sub4 = df_clstr['key_features'].copy()

        plot_clusters(num_clusters = num_clusters, 
                      feature_matrix = tfidf_feature_matrix,
                      cluster_data = cluster_data, 
                      movie_data = df_cluster,
                      plot_size = (12, 9))
```
